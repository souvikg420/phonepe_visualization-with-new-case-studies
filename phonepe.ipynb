```python
import os
import json
import pandas as pd
import mysql.connector
```

```python
#agg_insuarance
path1="C:/Users/User/OneDrive/Desktop/phonpe project/pulse/data/aggregated/insurance/country/india/state"
agg_ins_list=os.listdir(path1)
column1={"States":[],"Year":[],"Quarter":[],"Transaction_Type":[],"Transaction_Count":[],"Transaction_Amount":[]}
for states in agg_ins_list:
    curr_state=os.path.join(path1,states)
    agg_year_list=os.listdir(curr_state)
  
    for year in agg_year_list:
        curr_year=os.path.join(curr_state,year)
        agg_quartr_list=os.listdir(curr_year)

        for quarter in agg_quartr_list:
            curr_quarter=os.path.join(curr_year,quarter)
            data=open(curr_quarter,'r')
            A=json.load(data)
            for i in A["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                column1["States"].append(states)
                column1["Year"].append(year)
                column1["Quarter"].append(int(quarter.strip('.json')))
                column1["Transaction_Type"].append(name)
                column1["Transaction_Count"].append(count)
                column1["Transaction_Amount"].append(amount)

        
```

```python
#agg_insurance_table
aggre_insurance=pd.DataFrame(column1)
aggre_insurance["States"] = aggre_insurance["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggre_insurance["States"] = aggre_insurance["States"].str.replace("-"," ")
aggre_insurance["States"] = aggre_insurance["States"].str.title()
aggre_insurance['States'] = aggre_insurance['States'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
```

```python
#agg_user
path3="C:/Users/User/OneDrive/Desktop/phonpe project/pulse/data/aggregated/user/country/india/state"
agg_user_list=os.listdir(path3)
column3={"States":[],"Year":[],"Quarter":[],"Brand":[],"Transaction_Count":[],"Percentage":[]}
for state in agg_user_list:
    curr_states=os.path.join(path3,state)
    agg_year=os.listdir(curr_states)

    for year in agg_year:
        curr_year=os.path.join(curr_states,year)
        agg_quarter=os.listdir(curr_year)

        for quarter in agg_quarter:
            curr_quarter=os.path.join(curr_year,quarter)
            data=open(curr_quarter,"r")
            C=json.load(data)
            
            if C["data"]["usersByDevice"] is not None:
                for i in C["data"]["usersByDevice"]:
                    brand=i["brand"]
                    count=i["count"]
                    percentage=i["percentage"]
                    column3["States"].append(state)
                    column3["Year"].append(year)
                    column3["Brand"].append(brand)
                    column3["Percentage"].append(percentage)
                    column3["Transaction_Count"].append(count)
                    column3["Quarter"].append(int(quarter.strip(".json")))

```
